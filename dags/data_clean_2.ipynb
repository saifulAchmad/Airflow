```python
import pandas as pd 

df = pd.read_csv("~/airflow/Software Engineer Salaries.csv")
df.head(5)
```

```python
df [['min_range','max_range']]= df['Salary'].str.split(' - ',expand=True)
df.head(5)
```

```python
df [['max_range','source']]= df['max_range'].str.split('(', expand=True)
df.head(5)
```

```python
df['source'] = df['source'].str.replace('est.)', '', regex=False)
df['Date'] = df['Date'].str.replace('d', '', regex=False)
df.head(3)
```

```python
df=df.drop('Salary',axis=1)
df.head(3)
```

```python
df.to_json('result.json',orient='records')
```
